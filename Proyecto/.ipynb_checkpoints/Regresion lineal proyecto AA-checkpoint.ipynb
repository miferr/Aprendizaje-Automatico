# Regresion lineal

```python
from scipy.optimize import minimize

def hip(X, theta):
    return np.dot(X, theta)

def cost(theta, X, Y, reg):
    filas = np.shape(X)[0]
    m = np.shape(X)[0]
    return (1 / (2 * filas)) * np.sum(((hip(X, theta) - Y) ** 2)) + ((reg / (2 * filas)) * np.sum(theta[1:] ** 2))          

def gradient(theta, X, Y, reg):
    filas = np.shape(X)[0]
    gradiente = (1 / filas) * np.dot((hip(X, theta) - Y), X)
    result = np.zeros(np.shape(X)[1])
    result[0] = gradiente[0]
    result[1:] = gradiente[1:] + (theta[1:] * (reg / filas))
    return  result

# Ecuacion normal
def ecuacion_normal(X, Y):
    filas = np.shape(X)[0]
    th = np.linalg.inv(np.transpose(X).dot(X)).dot(np.transpose(X)).dot(Y)
    return th

def porcentaje(X, Y, thetas):
    ok = 0
    i = 0
    for h in hip(X, thetas):
        if h >= 0.5:
            if Y[i] == 1.0:
                ok +=1
        else:
            if Y[i] == 0.0:
                ok +=1
        i +=1
    return (ok / np.shape(X)[0])
```

Vamos a hacer la regresion lineal con la ecuacion normal

```python
thetas = ecuacion_normal(x_train, y_train)
print(porcentaje(x_test,y_test, thetas))
```

Regresion lineal descenso de gradiente

```python
lamdas = [0, 0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1, 3, 10, 100]
for reg in lamdas:
    theta = np.array([1] * (xn_train.shape[1]))
    theta_opt = minimize(fun=cost, x0=theta, args=(xn_train, y_train, reg), jac=gradient).x
    print(porcentaje(xn_test, y_test, theta_opt))
```

```python
def lamdas_error(n, training_y, crossval_y):
    plt.plot(n, training_y, label='Training error')
    plt.plot(n, crossval_y, label='Cross Validation error')
    plt.xlabel('Lambda')
    plt.ylabel('Error')
    plt.title('Errors for different lamda values')
    plt.legend()
    plt.show()

lamdas = [0, 0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1, 3, 10, 1000, 100000, 10000000]
X_error = []
Val_error = []
Total_error = []
#   Seleccionamos el lambda que da menor error entre los ejemplos de entrenamiento
# y los datos de validacion
for e in lamdas:
    theta_opt = minimize(fun=cost, x0=theta, args=(xn_train, y_train, e), jac=gradient).x
    Xe = cost(theta_opt, xn_train, y_train, e)
    Ve = cost(theta_opt, xn_val, y_val, e)
    X_error.append(Xe)
    Val_error.append(Ve)
    Total_error.append(np.abs(Xe - Ve))
    print('lamda: ', e,' Error: ', Xe,'Validation error: ', Ve)
 
i = np.argmin(Total_error)
min_lambda = lamdas[i];
print("Menor lambda: ", min_lambda) 

lamdas_error(lamdas, X_error, Val_error)
```

# Kernel reshulon

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

#def best_C_sigma(X, y_test, X_test, y_test):
values = np.linspace(0.1,10000, 100, dtype=float)
best_C = 0
best_sigma = 0
best_acc = 0
for C in values:
    for sigma in values:
        svm = SVC(kernel='rbf', C=C, gamma=1 / (2 * sigma**2))
        svm.fit(xn_train, y_train)
        acc = accuracy_score(y_val, svm.predict(xn_val))
        print(C,sigma, "acuraccy: ", acc)
        if acc > best_acc:
            best_C = C
            best_sigma = sigma
            best_acc = acc
```

```python
print(best_acc, best_C, best_sigma)
```
