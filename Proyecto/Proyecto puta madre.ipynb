```python
import numpy as np
import pandas as pd

#   Cargamos los datos
datos = pd.read_csv("water_potability.csv")
means = datos.mean(axis=0, skipna=True)

for e in datos:
    datos[e] = datos[e].fillna(means[e])
    
print(datos.head(15)) 

#Hasta aqui datos cargados y sustituimos los nan por la media
```
